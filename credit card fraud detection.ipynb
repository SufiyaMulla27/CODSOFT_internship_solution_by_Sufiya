```python
Delta_Time = pd.to_timedelta(ds['Time'], unit='s')
#Create derived columns Mins and hours
ds['Time_Hour'] = (Delta_Time.dt.components.hours).astype(int) # time of the hour 
ds['Time_Min'] = (Delta_Time.dt.components.minutes).astype(int) #minutue
```

```python
ds.drop('Time', axis = 1, inplace= True) #droping the Time column because now we have time in hr, min
```

```python
#plotting the time variatios in fraud transactions
plt.figure(figsize=(10,5))
sns.distplot(ds[ds['Class'] == 1]['Time_Hour'],kde=True,rug=False)
plt.title('Fraud Transactions by Hours')
```

```python
#normalising the Amount columns
from sklearn.preprocessing import StandardScaler
ds['normalizedAmount'] = StandardScaler().fit_transform(ds['Amount'].values.reshape(-1,1))
df = ds.drop(['Amount'],axis=1)  #droping the Amount column because we have normalisedAmount 
```

```python
df.head()
```

```python
#taking the features and the target value
x=df.drop(['Class'],axis=1)
y=df['Class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=10) #splitting the dataset into traing and testing
```

```python
from sklearn.tree import DecisionTreeClassifier #importing the model
```

```python
model =  DecisionTreeClassifier()
```

```python
model.fit(x_train, y_train) #training the model
```

```python
pred = model.predict(x_test) #predicting
```

```python
#checking the accuracy
acc = round(model.score(x_train,y_train)*100,2)
acc
```
